```python
from syfthub_sdk import SyftHubClient
import os

# 1. Adding or updating an environment variable
os.environ["SYFTHUB_URL"] = "https://syfthub.openmined.org"

with SyftHubClient() as client:
    user = client.auth.login(username="ionesiotest", password="password123")

    response = client.chat.complete(
            prompt="Is better to use less data during machine learning training?",
            model="ionesiotest/free-legal-assistant",
            data_sources=["ionesiotest/general-knowledge"],
            top_k=5,
            max_tokens=1024,
            temperature=0.7,
        )
```

```python
print(response.response)
print("-" * 40)
print(f"\nSources: {[s.path for s in response.sources]}")
print(f"Total time: {response.metadata.total_time_ms}ms")
```
